## Log(1+x) Transformation( as it has zeroes)

```python
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
import numpy as np
df['co_log']=np.log1p(df['vehicle_CO'])
plot_data(df,'co_log')
```

#### Reciprocal Transformation

```python
##### Square Root Transformation
df['co_sqaure']=df.vehicle_CO**(1/2)
plot_data(df,'co_sqaure')
```

```python
#### Exponential Transdormation
df['co_exponential']=df.vehicle_CO**(1/1.2)
plot_data(df,'co_exponential')
```
